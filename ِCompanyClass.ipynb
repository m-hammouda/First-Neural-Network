```python
import tensorflow as tf
import numpy as np
from tensorflow import keras

# ----------------------
# 1. تحضير البيانات
# ----------------------
X = np.array([
    [300000, -20000, 0.80, -0.06],
    [600000, 15000, 0.60, 0.02],
    [1200000, 120000, 0.30, 0.10],
    [800000, 40000, 0.50, 0.05],
    [2000000, 300000, 0.20, 0.15]
], dtype=float)

y = np.array([0, 1, 2, 1, 2])  # classes

# Normalisation
X = X / np.max(np.abs(X), axis=0)

# ----------------------
# 2. إنشاء النموذج
# ----------------------
model = tf.keras.Sequential([
    keras.layers.Dense(8, activation='relu', input_shape=[4]),
    keras.layers.Dense(3, activation='softmax')
])

# ----------------------
# 3. تجميع النموذج
# ----------------------
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# ----------------------
# 4. تدريب النموذج
# ----------------------
model.fit(X, y, epochs=600, verbose=0)

# ----------------------
# 5. تصنيف شركة جديدة
# ----------------------
new_company = np.array([[700000, 30000, 0.55, 0.04]], dtype=float)
new_company = new_company / np.max(np.abs(X), axis=0)

prediction = model.predict(new_company)
predicted_class = np.argmax(prediction)

print("التصنيف المتوقع:", predicted_class)
```
