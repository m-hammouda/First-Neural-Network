```python
import tensorflow as tf
import numpy as np
from tensorflow import keras

X = np.array([
    [600, 1, 1, 60],
    [3000, 4, 3, 10],
    [1500, 3, 2, 30],
    [2000, 4, 3, 20],
    [800, 2, 1, 50]
], dtype=float)

# المتغير الهدف (Price)
y = np.array([150000, 450000, 250000, 350000, 180000], dtype=float)

# ----------------------
# 2. إنشاء النموذج
# ----------------------
model = tf.keras.Sequential([
    keras.layers.Dense(units=1, input_shape=[4])  # 4 سمات لكل منزل
])

# ----------------------
# 3. تجميع النموذج
# ----------------------
model.compile(optimizer='sgd', loss='mean_squared_error')

# ----------------------
# 4. تدريب النموذج
# ----------------------
model.fit(X, y, epochs=1000, verbose=0)

# ----------------------
# 5. التنبؤ بسعر منزل جديد
# ----------------------
new_home = np.array([[1200, 3, 2, 25]])  # مثال منزل جديد
predicted_price = model.predict(new_home)
print(f"السعر المتوقع للمنزل الجديد: {predicted_price[0][0]:.2f} دولار")
```
