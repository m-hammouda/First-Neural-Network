```python
import tensorflow as tf
import numpy as np
from tensorflow import keras

# ----------------------
# 1. تحضير البيانات
# ----------------------
X = np.array([
    [500000, 10, 60000, 20000],
    [1200000, 25, 120000, 50000],
    [800000, 15, 80000, 30000],
    [2000000, 40, 200000, 90000],
    [300000, 6, 40000, 15000]
], dtype=float)

y = np.array([180000, 420000, 260000, 720000, 120000], dtype=float)

# تقييس البيانات لتفادي loss = nan
X = X / np.max(X, axis=0)
y = y / np.max(y)

# ----------------------
# 2. إنشاء النموذج
# ----------------------
model = tf.keras.Sequential([
    keras.layers.Dense(units=1, input_shape=[4])
])

# ----------------------
# 3. تجميع النموذج
# ----------------------
model.compile(
    optimizer='adam',
    loss='mean_squared_error'
)

# ----------------------
# 4. تدريب النموذج
# ----------------------
model.fit(X, y, epochs=800, verbose=0)

# ----------------------
# 5. التنبؤ بمصاريف شركة جديدة
# ----------------------
new_company = np.array([[900000, 18, 90000, 35000]], dtype=float)
new_company = new_company / np.max(X, axis=0)

prediction = model.predict(new_company)

print("المصاريف التشغيلية المتوقعة (تقريبية):",
      prediction[0][0] * np.max(y), "دينار")
```
